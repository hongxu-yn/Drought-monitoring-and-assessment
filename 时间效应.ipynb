```python
import xarray as xr
import numpy as np

nc_fn = project_dir / "data/ndvi/month/ndvi_month_2001_2025.nc"

out_dir = project_dir / "data/sai"
out_dir.mkdir(parents=True, exist_ok=True)
out_dir = out_dir / "month"
out_dir.mkdir(parents=True, exist_ok=True)

out_fn = out_dir / "sai_month_2001_2025.nc"

ds = xr.open_dataset(nc_fn)

var_name = None
for var in ds.data_vars:
    if "time" in ds[var].dims and var != "spatial_ref":
        var_name = var
        break

if var_name is None and "ndvi" in ds.data_vars:
    var_name = "ndvi"
elif var_name is None:
    for var in ds.data_vars:
        if var != "spatial_ref":
            var_name = var
            break

if var_name is None:
    raise ValueError("Could not find a suitable data variable.")

da = ds[var_name]

if not np.issubdtype(da["time"].dtype, np.datetime64):
    ds = xr.decode_cf(ds)
    da = ds[var_name]

def detrend_1d(y):
    y = np.asarray(y, dtype=float)
    mask = np.isfinite(y)

    if mask.sum() < 5:
        return np.full_like(y, np.nan)

    x = np.arange(y.size, dtype=float)
    p = np.polyfit(x[mask], y[mask], 1)
    trend = np.polyval(p, x)
    return y - trend

def detrend_monthly(da):
    out = []

    for m in range(1, 13):
        sub = da.where(da["time.month"] == m, drop=True)

        detrended = xr.apply_ufunc(
            detrend_1d,
            sub,
            input_core_dims=[["time"]],
            output_core_dims=[["time"]],
            vectorize=True,
            dask="parallelized",
            output_dtypes=[float],
        )

        detrended = detrended.assign_coords(time=sub.time)
        out.append(detrended)

    return xr.concat(out, dim="time").sortby("time")

da_detrended = detrend_monthly(da)
da_detrended.name = f"{var_name}_detrended"
da_detrended.attrs.update(da.attrs)
da_detrended.attrs["description"] = "Monthly detrended NDVI by calendar month"

ds_out = da_detrended.to_dataset()
ds_out.to_netcdf(out_fn)

print("处理完成！")
print(out_fn)
```

```python
#季节和年尺度ndvi去趋势
import xarray as xr
import pandas as pd
import numpy as np
import warnings
from numpy.polynomial.polyutils import RankWarning
def get_time_dim(da):
    for d in da.dims:
        if "time" in d.lower():
            return d
    raise ValueError(f"未找到时间维，当前 dims={da.dims}")

def detrend_1d(y):
    y = np.asarray(y, dtype=float)
    mask = np.isfinite(y)
    if mask.sum() < 3:
        return y
    if np.nanstd(y[mask]) < 1e-6:
        return y
    x = np.arange(mask.sum(), dtype=float)
    yy = y[mask]
    with warnings.catch_warnings():
        warnings.simplefilter("ignore", RankWarning)
        coef = np.polyfit(x, yy, 1)
    trend = np.polyval(coef, x)
    out = y.copy()
    out[mask] = yy - trend + np.nanmean(yy)
    return out

def detrend_data(da, dim="time"):
    return xr.apply_ufunc(
        detrend_1d,
        da,
        input_core_dims=[[dim]],
        output_core_dims=[[dim]],
        vectorize=True,
        dask="parallelized",
        output_dtypes=[float]
    )

project_dir = Path('/content/drive/MyDrive/Drought-monitoring-and-assessment')
nc_fn = project_dir / "data/ndvi/ndvi_month_2001_2025.nc"
output_dir = project_dir / "data/sai"
output_dir.mkdir(parents=True, exist_ok=True)

ds = xr.open_dataset(nc_fn)
time_dim = get_time_dim(da)
month = da[time_dim].dt.month
year = da[time_dim].dt.year
da = da.assign_coords(season_year=xr.where(month == 12, year + 1, year))

annual = da.resample({time_dim: "YS"}).mean()

spring = da.where(month.isin([3, 4, 5]), drop=True).groupby("season_year").mean(time_dim)
summer = da.where(month.isin([6, 7, 8]), drop=True).groupby("season_year").mean(time_dim)
autumn = da.where(month.isin([9, 10, 11]), drop=True).groupby("season_year").mean(time_dim)
winter = da.where(month.isin([12, 1, 2]), drop=True).groupby("season_year").mean(time_dim)

def finalize_and_detrend(da):
    if "season_year" in da.dims:
        da = da.assign_coords(
            season_year=pd.to_datetime(da["season_year"].values.astype(int), format="%Y")
        ).rename({"season_year": "time"})
    elif time_dim != "time":
        da = da.rename({time_dim: "time"})
    return detrend_data(da, dim="time")

spring_detrend = finalize_and_detrend(spring)
summer_detrend = finalize_and_detrend(summer)
autumn_detrend = finalize_and_detrend(autumn)
annual_detrend = finalize_and_detrend(annual)

winter_detrend = finalize_and_detrend(winter)
winter_detrend = winter_detrend.assign_coords(
    time=pd.to_datetime((winter_detrend["time"].dt.year.values - 1).astype(str), format="%Y")
)
winter_detrend = winter_detrend.sel(time=slice("2001-01-01", "2024-01-01"))

spring_detrend.to_netcdf(output_dir / "sai_spring_2001_2025.nc")
summer_detrend.to_netcdf(output_dir / "sai_summer_2001_2025.nc")
autumn_detrend.to_netcdf(output_dir / "sai_autumn_2001_2025.nc")
winter_detrend.to_netcdf(output_dir / "sai_winter_2001_2024.nc")
annual_detrend.to_netcdf(output_dir / "sai_year_2001_2025.nc")

print(f"处理完成！去趋势后的数据已保存至: {output_dir}")
```

# SPEI 格点化

```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import cdist
import xarray as xr
import os
from pathlib import Path
project_dir= Path('/content/drive/MyDrive/Drought-monitoring-and-assessment')
path_station = project_dir/ 'data/站点编号及信息_选定_1969-2025.xlsx'
path_spei = project_dir / 'data/spei/SPEI_month_1969_2025.csv'


df_stations = pd.read_excel(path_station)
spei_df = pd.read_csv(path_spei)
spei_df = spei_df[spei_df['year'].between(2001, 2026)]
combined_df = pd.merge(spei_df, df_stations[['station', 'lat', 'lon']], on='station', how='inner')

# 4. 定义插值网格 (20-30N, 97-107E, Res: 0.05)
res = 0.05
lon_coords = np.arange(97.0, 107.0 + res, res)
lat_coords = np.arange(20.0, 30.0 + res, res)
grid_lon, grid_lat = np.meshgrid(lon_coords, lat_coords)
grid_points = np.column_stack((grid_lon.ravel(), grid_lat.ravel()))


def idw_func(x, y, z, target_pts, p=2):
    known_pts = np.column_stack((x, y))
    dist = cdist(target_pts, known_pts)
    dist = np.where(dist == 0, 1e-10, dist)
    weights = 1.0 / (dist ** p)
    return np.sum(weights * z, axis=1) / np.sum(weights, axis=1)

combined_df['time'] = pd.to_datetime(combined_df[['year', 'month']].assign(day=1))
time_steps = sorted(combined_df['time'].unique())
results = np.zeros((len(time_steps), len(lat_coords), len(lon_coords)))

print(f"开始插值，共 {len(time_steps)} 个月份...")

# 外层循环：支持未来批量处理多个尺度，例如 [1, 3, 6, 12]
for time_scale in range (1,13):
    field_name = f'SPEI{time_scale}_Month'
    output_nc = f'/content/drive/MyDrive/Drought-monitoring-and-assessment/data/spei/SPEI{time_scale:02d}_month_2001_2025.nc'

    results = np.full((len(time_steps), len(lat_coords), len(lon_coords)), np.nan)

    for i, t in enumerate(time_steps):
        curr_df = combined_df[combined_df['time'] == t].dropna(subset=[field_name, 'lat', 'lon'])

        if not curr_df.empty:
            z_grid = idw_func(curr_df['lon'].values, curr_df['lat'].values,
                              curr_df[field_name].values, grid_points)

            results[i, :, :] = z_grid.reshape(len(lat_coords), len(lon_coords))
        else:
            print(f"警告：{pd.to_datetime(t).strftime('%Y-%m')} 缺失全部站点数据，记为 NaN")

        if (i+1) % 120 == 0:
            print(f"进度 ({field_name}): 已完成 {pd.to_datetime(t).year} 年数据")

    ds = xr.Dataset(
        {"spei": (["time", "lat", "lon"], results)},
        coords={"time": time_steps, "lat": lat_coords, "lon": lon_coords}
    )

    ds.spei.attrs["units"] = "1"
    ds.spei.attrs["long_name"] = f"Standardized Precipitation Evapotranspiration Index ({time_scale}-month)"
    ds.spei.attrs["missing_value"] = np.nan

    ds.to_netcdf(output_nc)
    print(f"[{field_name}] 插值完成！文件已保存至: {output_nc}\n" + "-"*40)
```

# 季节性和全年干旱与植被NDVI的相关关系

```python
from pathlib import Path
import xarray as xr
import numpy as np
from scipy.stats import pearsonr
from tqdm import tqdm

project_dir = Path('/content/drive/MyDrive/Drought-monitoring-and-assessment')
spei_dir = project_dir / "data/spei/year"
ndvi_dir = project_dir / "data/sai"
out_dir = project_dir / "output/corr_xy"
out_dir.mkdir(parents=True, exist_ok=True)

seasons = ["spring", "summer", "autumn", "winter", "year"]

def get_grid_var(ds):
    for name, da in ds.data_vars.items():
        if "time" in da.dims and len(da.dims) >= 3:
            return da
    raise ValueError(f"没有找到时空栅格变量，可选变量: {list(ds.data_vars)}")

def get_xy_dims(da):
    dims = [d for d in da.dims if d != "time"]
    if len(dims) != 2:
        raise ValueError(f"无法识别空间维度，当前 dims={da.dims}")
    return dims[0], dims[1]

for season in seasons:
    if season == "winter":
        ndvi_fn = ndvi_dir / "sai_winter_2001_2024.nc"
        spei_fn = spei_dir / "spei_winter_2001_2024.nc"
    else:
        ndvi_fn = ndvi_dir / f"sai_{season}_2001_2025.nc"
        spei_fn = spei_dir / f"spei_{season}_2001_2025.nc"

    ds_ndvi = xr.open_dataset(ndvi_fn)
    ds_spei = xr.open_dataset(spei_fn)

    ndvi = get_grid_var(ds_ndvi)
    spei = get_grid_var(ds_spei)

    ndvi, spei = xr.align(ndvi, spei, join="inner")

    y_dim, x_dim = get_xy_dims(ndvi)
    y = ndvi[y_dim]
    x = ndvi[x_dim]

    r_arr = np.full((len(y), len(x)), np.nan, dtype=np.float32)
    p_arr = np.full((len(y), len(x)), np.nan, dtype=np.float32)

    total = len(y) * len(x)
    with tqdm(total=total, desc=season) as pbar:
        for i in range(len(y)):
            for j in range(len(x)):
                x_ts = ndvi.isel({y_dim: i, x_dim: j}).values
                y_ts = spei.isel({y_dim: i, x_dim: j}).values

                mask = np.isfinite(x_ts) & np.isfinite(y_ts)
                if mask.sum() >= 3:
                    r, p = pearsonr(x_ts[mask], y_ts[mask])
                    r_arr[i, j] = r
                    p_arr[i, j] = p

                pbar.update(1)

    ds_out = xr.Dataset(
        {
            "r": ((y_dim, x_dim), r_arr),
            "p": ((y_dim, x_dim), p_arr),
        },
        coords={y_dim: y, x_dim: x}
    )

    out_fn = out_dir / f"corr_sai_spei_{season}.nc"
    ds_out.to_netcdf(out_fn)
    print(f"{season} done -> {out_fn}")
```

# 累积效应

```python
from pathlib import Path
import numpy as np
import xarray as xr
from scipy.stats import pearsonr
from tqdm.auto import tqdm


x_dir = project_dir / "data/spei/month"
y_dir = project_dir / "data/sai/month"
out_dir = project_dir / "output/time_effect"
out_dir.mkdir(parents=True, exist_ok=True)

out_nc = out_dir / "cumulative_effect_spei_sai_2001_2024.nc"

def get_main_var(ds):
    for v in ds.data_vars:
        if v != "spatial_ref":
            return v
    raise ValueError("未找到有效数据变量")

def corr_p_1d(x, y):
    x = np.asarray(x, dtype=float)
    y = np.asarray(y, dtype=float)

    mask = np.isfinite(x) & np.isfinite(y)
    if mask.sum() < 3:
        return np.nan, np.nan

    x_valid = x[mask]
    y_valid = y[mask]

    if np.nanstd(x_valid) == 0 or np.nanstd(y_valid) == 0:
        return np.nan, np.nan

    r, p = pearsonr(x_valid, y_valid)
    return r, p

# 读取 SAI
y_nc = y_dir / "sai_month_2001_2025.nc"
ds_y = xr.open_dataset(y_nc)
ds_y = ds_y.sel(time=slice("2001-01-01", "2024-12-31"))

var_y = get_main_var(ds_y)
da_y = ds_y[var_y]

if not np.issubdtype(da_y["time"].dtype, np.datetime64):
    ds_y = xr.decode_cf(ds_y)
    da_y = ds_y[var_y]

da_y = da_y.transpose("time", "lat", "lon")

lat = da_y["lat"].values
lon = da_y["lon"].values

# 最终结果数组
best_r_arr = np.full((len(lat), len(lon)), np.nan, dtype=np.float32)
max_abs_r_arr = np.full((len(lat), len(lon)), np.nan, dtype=np.float32)
best_p_arr = np.full((len(lat), len(lon)), np.nan, dtype=np.float32)
best_scale_arr = np.full((len(lat), len(lon)), np.nan, dtype=np.float32)


for time_scale in range(1, 13):
    x_nc = x_dir / f"SPEI{time_scale:02d}_month_2001_2025.nc"
    print(f"\n正在处理: SPEI{time_scale:02d}")

    ds_x = xr.open_dataset(x_nc)
    var_x = get_main_var(ds_x)
    da_x = ds_x[var_x]

    if not np.issubdtype(da_x["time"].dtype, np.datetime64):
        ds_x = xr.decode_cf(ds_x)
        da_x = ds_x[var_x]

    da_x = da_x.sel(time=slice("2001-01-01", "2024-12-31"))
    da_x = da_x.transpose("time", "lat", "lon")

    da_x_aligned, da_y_aligned = xr.align(da_x, da_y, join="inner")

    if da_x_aligned.dims != ("time", "lat", "lon"):
        raise ValueError(f"SPEI{time_scale:02d} 维度异常: {da_x_aligned.dims}")
    if da_y_aligned.dims != ("time", "lat", "lon"):
        raise ValueError(f"SAI 维度异常: {da_y_aligned.dims}")

    x_vals = da_x_aligned.values
    y_vals = da_y_aligned.values

    r_arr = np.full((len(lat), len(lon)), np.nan, dtype=np.float32)
    p_arr = np.full((len(lat), len(lon)), np.nan, dtype=np.float32)

    for i in tqdm(range(len(lat)), desc=f"SPEI{time_scale:02d} 行进度"):
        for j in range(len(lon)):
            x = x_vals[:, i, j]
            y = y_vals[:, i, j]
            r, p = corr_p_1d(x, y)
            r_arr[i, j] = r
            p_arr[i, j] = p

    abs_r_arr = np.abs(r_arr)

    # 首次赋值：当前为空且本次有效
    init_mask = np.isnan(max_abs_r_arr) & np.isfinite(abs_r_arr)

    best_r_arr[init_mask] = r_arr[init_mask]
    max_abs_r_arr[init_mask] = abs_r_arr[init_mask]
    best_p_arr[init_mask] = p_arr[init_mask]
    best_scale_arr[init_mask] = time_scale

    # 更新：当前时间尺度的 |r| 更大
    update_mask = np.isfinite(abs_r_arr) & np.isfinite(max_abs_r_arr) & (abs_r_arr > max_abs_r_arr)

    best_r_arr[update_mask] = r_arr[update_mask]
    max_abs_r_arr[update_mask] = abs_r_arr[update_mask]
    best_p_arr[update_mask] = p_arr[update_mask]
    best_scale_arr[update_mask] = time_scale

best_sig_arr = (best_p_arr < 0.05).astype(np.int8)

ds_out = xr.Dataset(
    {
        "best_r": (("lat", "lon"), best_r_arr),
        "best_p": (("lat", "lon"), best_p_arr),
        "best_scale": (("lat", "lon"), best_scale_arr),
    },
    coords={
        "lat": lat,
        "lon": lon,
    },
)

ds_out["best_r"].attrs = {
    "long_name": "Signed Pearson correlation coefficient at the time scale with maximum absolute correlation",
    "description": "Original signed correlation coefficient corresponding to the maximum absolute correlation across SPEI01-SPEI12"
}
ds_out["best_p"].attrs = {
    "long_name": "P-value corresponding to the best time scale",
    "description": "P-value associated with the time scale yielding the maximum absolute correlation"
}
ds_out["best_scale"].attrs = {
    "long_name": "Best SPEI time scale",
    "description": "Time scale (1-12 months) corresponding to the maximum absolute correlation"
}

ds_out.attrs["time_period"] = "2001-2024"
ds_out.attrs["x_variable"] = "SPEI01-SPEI12"
ds_out.attrs["y_variable"] = var_y
ds_out.attrs["method"] = "Pixel-wise Pearson correlation analysis; final output retains only the time scale with the maximum absolute correlation"
ds_out.attrs["note"] = "If multiple time scales have the same absolute correlation, the first encountered scale is retained."

ds_out.to_netcdf(out_nc)

print("\n全部完成！")
print(out_nc)
```

```python
from pathlib import Path
import numpy as np
import xarray as xr
import rioxarray
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.gridspec as gridspec
from matplotlib.colors import BoundaryNorm, ListedColormap
from matplotlib import font_manager as fm
import cartopy.crs as ccrs
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter

# =========================
# 1. 配置与路径
# =========================
CONFIG = {
    "project_dir": Path("/content/drive/MyDrive/Drought-monitoring-and-assessment"),
    "shp_path": "/content/drive/MyDrive/gis/boua_s.shp",
    "font_zh": "/content/drive/MyDrive/fonts/SimSun.ttf",
    "font_en": "/content/drive/MyDrive/fonts/TimesNewRoman.ttf",
    "dpi": 400
}

def setup_fonts(config):
    if Path(config["font_zh"]).exists():
        fm.fontManager.addfont(config["font_zh"])
        prop_zh = fm.FontProperties(fname=config["font_zh"])
        mpl.rcParams['font.family'] = prop_zh.get_name()

    f_zh = fm.FontProperties(fname=config["font_zh"], size=12) if Path(config["font_zh"]).exists() else None
    f_en = fm.FontProperties(fname=config["font_en"], size=11) if Path(config["font_en"]).exists() else None
    mpl.rcParams["axes.unicode_minus"] = False
    return f_zh, f_en

font_zh, font_en = setup_fonts(CONFIG)

# =========================
# 2. 数据处理与统计函数
# =========================
def process_data(nc_path, gdf):
    ds = xr.open_dataset(nc_path).load()

    def clip_da(da):
        return (
            da.rio.set_spatial_dims(x_dim="lon", y_dim="lat")
              .rio.write_crs("EPSG:4326")
              .rio.clip(gdf.geometry, gdf.crs, drop=True)
        )

    best_scale = clip_da(ds["best_scale"])
    best_r     = clip_da(ds["best_r"])
    best_p     = clip_da(ds["best_p"])

    sig_mask = best_p < 0.05
    scale_sig = best_scale.where(sig_mask)
    r_sig     = best_r.where(sig_mask)
    bg_mask   = best_scale.where(best_scale.notnull()) * 0 + 1

    return scale_sig, r_sig, bg_mask, best_p, best_scale, best_r

def calculate_stats(da_full, da_p, type='scale'):
    valid_mask = da_full.notnull()
    total_valid = valid_mask.sum().values
    insig_count = (valid_mask & (da_p >= 0.05)).sum().values
    results = {"不显著": insig_count}

    if type == 'scale':
        bins = [(1, 3), (4, 6), (7, 9), (10, 12)]
        labels = ["1-3", "4-6", "7-9", "10-12"]
        for i, (low, high) in enumerate(bins):
            mask = (da_p < 0.05) & (da_full >= low) & (da_full <= high)
            results[labels[i]] = mask.sum().values
    else:
        r_bins = [(-1.1, -0.6), (-0.6, -0.3), (-0.3, 0), (0, 0.3), (0.3, 0.6), (0.6, 1.1)]
        r_labels = ["<-0.6", "-0.6~-0.3", "-0.3~0", "0~0.3", "0.3~0.6", ">0.6"]
        for i, (low, high) in enumerate(r_bins):
            mask = (da_p < 0.05) & (da_full > low) & (da_full <= high)
            results[r_labels[i]] = mask.sum().values

    return {k: (v / total_valid) * 100 for k, v in results.items()}

# =========================
# 3. 绘图辅助函数
# =========================
def setup_map_ax(ax, extent, gdf, f_en):
    ax.set_extent(extent, crs=ccrs.PlateCarree())
    gdf.boundary.plot(ax=ax, edgecolor="black", linewidth=0.7, zorder=4)

    ax.set_xticks([98, 100, 102, 104, 106], crs=ccrs.PlateCarree())
    ax.set_yticks([22, 24, 26, 28], crs=ccrs.PlateCarree())
    ax.xaxis.set_major_formatter(LongitudeFormatter())
    ax.yaxis.set_major_formatter(LatitudeFormatter())

    ax.set_xlabel(""); ax.set_ylabel("")
    for t in ax.get_xticklabels() + ax.get_yticklabels():
        t.set_fontproperties(f_en)
        t.set_fontsize(10)

# =========================
# 4. 主绘图流程
# =========================
plt.close('all')

# 数据准备
gdf = gpd.read_file(CONFIG["shp_path"])
if gdf.crs is None: gdf.set_crs("EPSG:4326", inplace=True)

nc_fn = CONFIG["project_dir"] / "output/time_effect/cumulative_effect_spei_sai_2001_2024.nc"
scale_sig, r_sig, bg_mask, p_clip, scale_full, r_full = process_data(nc_fn, gdf)

stats_scale = calculate_stats(scale_full, p_clip, type='scale')
stats_r = calculate_stats(r_full, p_clip, type='r')

# 坐标与颜色设置
lon, lat = scale_sig.lon.values, scale_sig.lat.values
extent = [lon.min(), lon.max(), lat.min(), lat.max()]
cmap_bg = ListedColormap(["#eeeeee"])
cmap_scale = plt.get_cmap("Spectral_r", 12)
norm_scale = BoundaryNorm(np.arange(0.5, 13.5, 1), cmap_scale.N)
cmap_r = plt.get_cmap("RdBu_r")

# 创建画布 - 增加横向间距 wspace 确保图例不重叠
fig = plt.figure(figsize=(11, 8))
gs = gridspec.GridSpec(2, 2, width_ratios=[1.2, 1], wspace=0.3, hspace=0.2)

# --- 第一行：响应时间尺度 ---
ax1 = fig.add_subplot(gs[0, 0], projection=ccrs.PlateCarree())
setup_map_ax(ax1, extent, gdf, font_en)
ax1.pcolormesh(lon, lat, bg_mask.values, cmap=cmap_bg, transform=ccrs.PlateCarree(), zorder=1)
im1 = ax1.pcolormesh(lon, lat, scale_sig.values, cmap=cmap_scale, norm=norm_scale, transform=ccrs.PlateCarree(), zorder=2)

# 垂直图例放在右侧
cbar1 = fig.colorbar(im1, ax=ax1, orientation="vertical", pad=0.03, shrink=0.8, aspect=20)
cbar1.set_label("Time scale (month)", fontproperties=font_en, labelpad=10)
cbar1.set_ticks(range(1, 13))

ax2 = fig.add_subplot(gs[0, 1])
scale_colors = ['#d9d9d9'] + [cmap_scale(x) for x in np.linspace(0.1, 0.9, 4)]
ax2.bar(stats_scale.keys(), stats_scale.values(), color="lightgray", edgecolor='black', linewidth=0.6)


ax2.set_ylabel("占比 (%)", fontproperties=font_zh)

# --- 第二行：响应强度 ---
ax3 = fig.add_subplot(gs[1, 0], projection=ccrs.PlateCarree())
setup_map_ax(ax3, extent, gdf, font_en)
ax3.pcolormesh(lon, lat, bg_mask.values, cmap=cmap_bg, transform=ccrs.PlateCarree(), zorder=1)
im2 = ax3.pcolormesh(lon, lat, r_sig.values, cmap=cmap_r, vmin=-0.8, vmax=0.8, transform=ccrs.PlateCarree(), zorder=2)


# 垂直图例放在右侧
cbar2 = fig.colorbar(im2, ax=ax3, orientation="vertical", pad=0.03, shrink=0.8, aspect=20)
cbar2.set_label("Correlation coefficient (r)", fontproperties=font_en, labelpad=10)

ax4 = fig.add_subplot(gs[1, 1])
r_colors = ['#d9d9d9'] + [cmap_r(x) for x in [0.1, 0.25, 0.4, 0.6, 0.75, 0.9]]
ax4.bar(stats_r.keys(), stats_r.values(), color="lightgray", edgecolor='black', linewidth=0.6)

ax4.set_ylabel("占比 (%)", fontproperties=font_zh)

# --- 统一修饰统计图 ---
for ax in [ax2, ax4]:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for rect in ax.patches:
        h = rect.get_height()
        ax.annotate(f'{h:.1f}%', xy=(rect.get_x() + rect.get_width()/2, h),
                    xytext=(0, 3), textcoords="offset points", ha='center', fontproperties=font_en, fontsize=9)
    for t in ax.get_xticklabels():
        text = t.get_text()
        t.set_fontproperties(font_zh if "不显著" in text else font_en)
        t.set_fontsize(10)
    for t in ax.get_yticklabels():
        t.set_fontproperties(font_en)
fig.canvas.draw()

# 第一行：ax1 + ax2 的整体标题
pos1 = ax1.get_position()
pos2 = ax2.get_position()
x_center_row1 = (pos1.x0 + pos2.x1) / 2
y_row1 = max(pos1.y1, pos2.y1) + 0.015

fig.text(
    x_center_row1, y_row1,
    "(a) 响应时间",
    ha="center", va="bottom",
    fontproperties=font_zh,
    fontsize=14
)

# 第二行：ax3 + ax4 的整体标题
pos3 = ax3.get_position()
pos4 = ax4.get_position()
x_center_row2 = (pos3.x0 + pos4.x1) / 2
y_row2 = max(pos3.y1, pos4.y1) + 0.015

fig.text(
    x_center_row2, y_row2,
    "(b) 响应强度",
    ha="center", va="bottom",
    fontproperties=font_zh,
    fontsize=14
)
plt.setp(ax4.get_xticklabels(), rotation=20, ha='right')

plt.savefig(CONFIG["project_dir"] / "figures/cumulative_effect_matrix_final_v.png", dpi=CONFIG["dpi"], bbox_inches="tight")
plt.show()
```

# 滞后效应

```python
from pathlib import Path
import numpy as np
import xarray as xr
from scipy.stats import pearsonr
from tqdm.auto import tqdm

x_dir = project_dir / "data/spei/month"
y_dir = project_dir / "data/sai/month"
out_dir = project_dir / "output/time_effect"
out_dir.mkdir(parents=True, exist_ok=True)


spei_scale = 3
max_lag = 12

out_nc = out_dir / f"lag_effect_spei{spei_scale:02d}_sai_2001_2024.nc"

def get_main_var(ds):
    for v in ds.data_vars:
        if v != "spatial_ref":
            return v
    raise ValueError("未找到有效数据变量")

def corr_p_1d(x, y):
    x = np.asarray(x, dtype=float)
    y = np.asarray(y, dtype=float)

    mask = np.isfinite(x) & np.isfinite(y)
    if mask.sum() < 3:
        return np.nan, np.nan

    x_valid = x[mask]
    y_valid = y[mask]

    if np.nanstd(x_valid) == 0 or np.nanstd(y_valid) == 0:
        return np.nan, np.nan

    r, p = pearsonr(x_valid, y_valid)
    return r, p

# 读取 SAI
y_nc = y_dir / "sai_month_2001_2025.nc"
ds_y = xr.open_dataset(y_nc)
ds_y = ds_y.sel(time=slice("2001-01-01", "2024-12-31"))

var_y = get_main_var(ds_y)
da_y = ds_y[var_y]

if not np.issubdtype(da_y["time"].dtype, np.datetime64):
    ds_y = xr.decode_cf(ds_y)
    da_y = ds_y[var_y]

da_y = da_y.transpose("time", "lat", "lon")

# 读取 SPEI03
x_nc = x_dir / f"SPEI{spei_scale:02d}_month_2001_2025.nc"
ds_x = xr.open_dataset(x_nc)
ds_x = ds_x.sel(time=slice("2001-01-01", "2024-12-31"))

var_x = get_main_var(ds_x)
da_x = ds_x[var_x]

if not np.issubdtype(da_x["time"].dtype, np.datetime64):
    ds_x = xr.decode_cf(ds_x)
    da_x = ds_x[var_x]

da_x = da_x.transpose("time", "lat", "lon")

lat = da_y["lat"].values
lon = da_y["lon"].values

best_r_arr = np.full((len(lat), len(lon)), np.nan, dtype=np.float32)
best_p_arr = np.full((len(lat), len(lon)), np.nan, dtype=np.float32)
best_lag_arr = np.full((len(lat), len(lon)), np.nan, dtype=np.float32)
max_abs_r_arr = np.full((len(lat), len(lon)), np.nan, dtype=np.float32)

for lag in range(0, max_lag + 1):
    print(f"\n正在处理 lag = {lag}")

    # 当月 SAI 对应 前 lag 个月的 SPEI03
    da_x_lag = da_x.shift(time=lag)

    da_x_aligned, da_y_aligned = xr.align(da_x_lag, da_y, join="inner")

    x_vals = da_x_aligned.values
    y_vals = da_y_aligned.values

    r_arr = np.full((len(lat), len(lon)), np.nan, dtype=np.float32)
    p_arr = np.full((len(lat), len(lon)), np.nan, dtype=np.float32)

    for i in tqdm(range(len(lat)), desc=f"lag={lag} 行进度"):
        for j in range(len(lon)):
            x = x_vals[:, i, j]
            y = y_vals[:, i, j]
            r, p = corr_p_1d(x, y)
            r_arr[i, j] = r
            p_arr[i, j] = p

    abs_r_arr = np.abs(r_arr)

    init_mask = np.isnan(max_abs_r_arr) & np.isfinite(abs_r_arr)
    best_r_arr[init_mask] = r_arr[init_mask]
    best_p_arr[init_mask] = p_arr[init_mask]
    best_lag_arr[init_mask] = lag
    max_abs_r_arr[init_mask] = abs_r_arr[init_mask]

    update_mask = np.isfinite(abs_r_arr) & np.isfinite(max_abs_r_arr) & (abs_r_arr > max_abs_r_arr)
    best_r_arr[update_mask] = r_arr[update_mask]
    best_p_arr[update_mask] = p_arr[update_mask]
    best_lag_arr[update_mask] = lag
    max_abs_r_arr[update_mask] = abs_r_arr[update_mask]

best_sig_arr = (best_p_arr < 0.05).astype(np.int8)

ds_out = xr.Dataset(
    {
        "best_r": (("lat", "lon"), best_r_arr),
        "best_p": (("lat", "lon"), best_p_arr),
        "best_lag": (("lat", "lon"), best_lag_arr),
        "max_abs_r": (("lat", "lon"), max_abs_r_arr),
        "best_sig_p05": (("lat", "lon"), best_sig_arr),
    },
    coords={
        "lat": lat,
        "lon": lon,
    },
)

ds_out["best_r"].attrs = {
    "long_name": "Signed Pearson correlation coefficient at the best lag",
    "description": "Original signed correlation coefficient corresponding to the maximum absolute correlation across lag 0-12 months"
}
ds_out["best_p"].attrs = {
    "long_name": "P-value corresponding to the best lag",
    "description": "P-value associated with the lag yielding the maximum absolute correlation"
}
ds_out["best_lag"].attrs = {
    "long_name": "Best lag month",
    "description": "Lag time (0-12 months) corresponding to the maximum absolute correlation"
}
ds_out["max_abs_r"].attrs = {
    "long_name": "Maximum absolute Pearson correlation coefficient",
    "description": "Maximum absolute correlation across lag 0-12 months"
}
ds_out["best_sig_p05"].attrs = {
    "long_name": "Significance mask at p < 0.05 for the best lag",
    "description": "1 indicates significant correlation at p < 0.05; 0 indicates not significant"
}

ds_out.attrs["time_period"] = "2001-2024"
ds_out.attrs["x_variable"] = f"SPEI{spei_scale:02d}"
ds_out.attrs["y_variable"] = var_y
ds_out.attrs["method"] = "Pixel-wise lagged Pearson correlation analysis using SPEI03 and SAI"
ds_out.attrs["note"] = "The lag with the maximum absolute correlation was retained; if ties occur, the first encountered lag is retained."

ds_out.to_netcdf(out_nc)

print("\n全部完成！")
print(out_nc)
```
